## （一）
针对CreditMetrics课件(c)种的资产相关性一节的例子，给出ABC公司和XYZ公司的资产收益率之间的相关系数计算的详细过程和结果

设定
- 考虑ABC和XYZ两个公司
- 假设确定ABC属于美国的化工业，并且他的股权收益的波动性90%由美国的化工业指数来解释
- 假设XYZ公司的业务，占德国保险业的权重为75%，占德国银行业和金融业的权重为25%，其中系统性因素对收益的波动性的解释为80%

我们有这些底层因子的
- 波动率向量

|风险因素|波动率|
|--|--|
|美国化工业|2.03%|
|德国保险业|2.09%|
|德国银行业与金融业|1.25% |

- 相关性矩阵

||美国化工业|德国保险业|德国银行业与金融业|
|--|--|--|--|
|美国化工业|1|0.16|0.08 |
|德国保险业|0.16| 1|0.34|
|德国银行业与金融业|0.08 |0.34 | 1|






## 解答

在上述条件下，我们可以认为
$$
R^{ABC} =w_1r^{(USCm)}+w_2r^{(ABC)}
$$
ABC的波动性的90%由指数决定可得
$$
w_1 = 0.90,w_2 = \sqrt{1-w_1^2} = 0.44
$$


在上述条件下，我们可以认为
$$
R^{XYZ} =w_1r^{(DeIn)}+w_2r^{(DeBa)}+w_3r^{(XYZ)}
$$
先计算指数的波动性
$$
\hat{\sigma} = (0.75^2\sigma^2_{DeIn} + 0.25^2\sigma^2_{DeBa}+2 \times 0.75 \times 0.25 \rho(DeIn,DeBa)\sigma_{DeIn}\sigma_{DeBa}) = 0.017
$$

$$
w_1 = 0.8 \frac{0.75 \times \sigma_{Deln}}{\hat{\sigma}}  = 0.74 \\
w_2 = 0.8 \frac{0.75 \times \sigma_{DeBa}}{\hat{\sigma}}  = 0.15
$$

XYZ的波动性的80%由指数决定可得
$$
w_3 = \sqrt{1-0.8^2} = 0.6
$$


所以最终可以认为
$$
R^{ABC} =0.90 r^{(USCm)} + 0.44 r^{(ABC)} \\
R^{XYZ} =0.74r^{(DeIn)} + 0.15r^{(DeBa)}+ 0.15r^{(XYZ)} \\
$$

其中$r^{(USCm)}，r^{(DeIn)} ，r^{(DeBa)}$ 相关性如上表
$r^{(ABC)},r^{(XYZ)}$ 与上述3个指数相互独立,且相互之间也独立


最后计算

$$ 
\begin{align}
\rho(R^{ABC},R^{XYZ})  & =  0.90 \times 0.74 \rho(USC_m,DeIn)+0.90 \times 0.15\rho(USCm,DeBa) \\
& = 0.10656 + 0.0108\\
& = 0.11736
\end{align}
$$
