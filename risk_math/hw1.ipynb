```python
def get_data(comb_stock_codes):
    working_df = pd.read_csv('../data/csmar_close/TRD_Dalyr.txt',
                         sep='\t',
                         dtype={'Stkcd': str} )[['Trddt','Stkcd','Adjprcwd']]
    workding_df = working_df.set_index(['Trddt','Stkcd']).unstack('Stkcd')
    df = workding_df.T.dropna().T
    df.columns = df.columns.get_level_values(1)
    return df[comb_stock_codes] 
```

```python
import pandas as pd
import numpy as np
```

```python
comb_stock_codes = ['000001','002330','000607','000021']
```

```python

price_df = get_data(comb_stock_codes)
```
