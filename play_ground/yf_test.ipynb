```python
# 下载 XAU/USD 现货黄金 日线数据
import yfinance as yf
import pandas as pd

# ==================== 参数设置 ====================
ticker = "XAUUSD=X"           # 现货黄金 (XAU/USD)
start_date = "2023-01-01"     # 开始日期
end_date = "2025-11-11"       # 结束日期（含）
interval = "1d"               # 日线
proxy = "http://127.0.0.1:7890"  # Clash/V2Ray 代理（若无代理，可设为 None）

# ==================== 下载数据 ====================
print(f"正在下载 {ticker} 从 {start_date} 到 {end_date} 的日线数据...")

data = yf.download(
    tickers=ticker,
    start=start_date,
    end=end_date,
    interval=interval,
    threads=False,            # 推荐关闭多线程（代理更稳定）
    proxy=proxy if proxy else None,
    auto_adjust=True,         # 自动复权（推荐开启）
    progress=True
)

# ==================== 数据检查与保存 ====================
if data.empty:
    print("错误：未获取到数据！请检查网络、代理或日期。")
else:
    print(f"\n下载成功！共 {len(data)} 个交易日。")
    print(f"时间范围: {data.index[0].date()} → {data.index[-1].date()}")

    print("\n前 5 行数据：")
    print(data.head())

    print("\n后 5 行数据：")
    print(data.tail())

    # 保存为 CSV（带日期索引）
    filename = f"XAUUSD_daily_{start_date}_to_{end_date}.csv"
    data.to_csv(filename)
    print(f"\n数据已保存至：{filename}")

    # 可选：重置索引，方便 Excel 使用
    data_reset = data.reset_index()
    data_reset.to_csv(filename.replace('.csv', '_with_date.csv'), index=False)
    print(f"带日期列版本已保存：{filename.replace('.csv', '_with_date.csv')}")
```
